```python
# Dependencies
from bs4 import BeautifulSoup as bs
import pandas as pd
from splinter import Browser
import time
import re
```

```python
executable_path = {'executable_path': 'drivers\chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
headline,price,location,bed = [],[],[],[]
#create columns
craigslist_result = pd.DataFrame(columns=['HeadLine','Price','Location','Bed'])

url = "https://sfbay.craigslist.org/search/apa?sort=priceasc&min_price=1000&availabilityMode=0&sale_date=all+dates"
browser.visit(url)
soup = bs(browser.html, "html.parser")
#Get number of pages to scroll and divide by display per page , so that it goes through that many pages
total = int(soup.find(class_="totalcount").get_text())

word_check = 'Farm'

total = int(total/120)
print(total)
for j in range(1,total):
    
    soup = bs(browser.html, 'lxml')
    for i in soup.find_all("li", class_="result-row"):
        
        #Checking Headline
        
        xheadline = i.find(class_="result-title").get_text()
        
        #Checking if Headline has 'Farm'
        check_result = xheadline.find(word_check)
        
        #Appending to list if Headline doesn't have 'Farm'
        if check_result == -1:
            headline.append(xheadline)
            
            #Checking Price
            xprice = i.find(class_='result-price').get_text()
            xprice = xprice.replace('$', '').replace(',', ')') # removing $ and ','
            price.append(xprice)           
            
            #Checking Bedroom option, 
            #Data clean up , remove -, br, check if the value ha numeric and txt.
            try:
                xbed = i.find(class_='housing').get_text().strip().replace('-','').replace('br','').replace(' ','')
                
                check_numeric = xbed.isnumeric()
                
                if check_numeric == True:
                    bed.append(int(xbed))
                else:
                    bed.append(xbed)
            except:
                bed.append("NA")
            
            #Checking Location
            try:
                xlocation = i.find(class_="result-hood").get_text()
                xlocation = xlocation.replace('(','').replace(')','').replace('!','').replace('@','')\
                .replace('>','').replace('&','').replace('^','').replace('+','').replace('~','')\
                .replace('-','').replace('*','')# replace if special characters

                check_numeric = xlocation.isnumeric()
                
                if check_numeric == True: #Only Numeric 
                    location.append('NA')
                else:
                    location.append(xlocation)
            except:
                location.append('NA')
                        
    if j <= total:
        print(j)
        browser.click_link_by_partial_text('next')
        time.sleep(2)

print(len(headline))
print(len(price))
print(len(location))

my_dict = dict( Title = headline, Price = price,Location = location,Bed = bed) 
craigslist_result = pd.DataFrame.from_dict(my_dict, orient='index').T
craigslist_result.to_csv("craigs.csv")
```
